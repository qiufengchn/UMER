```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 读取完整数据
buildings_path = r"F:\ScientificDatabase\paper03\wuhan_district\buildings_with_landuse_full.shp"
buildings_df = gpd.read_file(buildings_path)

# 打印数据基本信息
print("数据字段:", list(buildings_df.columns))
print("数据量:", len(buildings_df))

# 计算基础数据
buildings_df['building_area'] = buildings_df.geometry.area  # 建筑底面积

# 按一级和二级用地类型分组计算
level1_stats = buildings_df.groupby(['Level1_cn']).agg({
    'building_area': 'sum',    # 建筑底面积总和
    'F_AREA': 'first',         # 用地面积
    'geometry': 'count'        # 建筑物数量
}).reset_index()

level2_stats = buildings_df.groupby(['Level1_cn', 'Level2_cn']).agg({
    'building_area': 'sum',
    'F_AREA': 'first',
    'geometry': 'count'
}).reset_index()

# 计算建筑密度和容积率
def calculate_metrics(df):
    df['building_density'] = (df['building_area'] / df['F_AREA'] * 100).round(2)
    # 根据不同用地类型设置不同的平均层数
    df['avg_floors'] = np.where(df['Level1_cn'].str.contains('居住'), 6,
                      np.where(df['Level1_cn'].str.contains('商业'), 4,
                      np.where(df['Level1_cn'].str.contains('工业'), 2, 3)))
    df['FAR'] = (df['building_area'] * df['avg_floors'] / df['F_AREA']).round(2)
    return df

level1_stats = calculate_metrics(level1_stats)
level2_stats = calculate_metrics(level2_stats)

# 显示结果
print("\n一级用地类型统计:")
print("================================")
print(level1_stats[['Level1_cn', 'geometry', 'building_density', 'FAR']])

# 绘制可视化图表
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 创建子图
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(20, 16))

# 1. 建筑密度柱状图
level1_stats.plot(kind='bar', x='Level1_cn', y='building_density', ax=ax1)
ax1.set_title('各类用地建筑密度 (%)')
ax1.set_xlabel('一级用地类型')
ax1.set_ylabel('建筑密度 (%)')
ax1.tick_params(axis='x', rotation=45)

# 2. 容积率柱状图
level1_stats.plot(kind='bar', x='Level1_cn', y='FAR', ax=ax2)
ax2.set_title('各类用地容积率')
ax2.set_xlabel('一级用地类型')
ax2.set_ylabel('容积率')
ax2.tick_params(axis='x', rotation=45)

# 3. 建筑密度空间分布
buildings_df.plot(column='Level1_cn', 
                 categorical=True,
                 legend=True,
                 ax=ax3,
                 legend_kwds={'bbox_to_anchor': (1.3, 1),
                            'title': '用地类型'})
ax3.set_title('建筑分布与用地类型')
ax3.axis('equal')

# 4. 热力图
buildings_df['centroid'] = buildings_df.geometry.centroid
buildings_df.plot(ax=ax4, 
                 alpha=0.1, 
                 color='grey')
ax4.scatter(buildings_df.centroid.x, 
           buildings_df.centroid.y,
           c=buildings_df['building_area'],
           alpha=0.5,
           cmap='YlOrRd',
           s=1)
ax4.set_title('建筑密度热力分布')
ax4.axis('equal')

plt.tight_layout()
plt.show()

# 导出统计结果到Excel
output_excel = r"F:\ScientificDatabase\paper03\wuhan_district\building_density_stats.xlsx"
with pd.ExcelWriter(output_excel) as writer:
    level1_stats.to_excel(writer, sheet_name='一级用地统计', index=False)
    level2_stats.to_excel(writer, sheet_name='二级用地统计', index=False)

print(f"\n统计结果已保存至: {output_excel}")
```

```python
SCD = ∑(建筑体积) / (网格面积 × 参考高度)
```

SCD = ∑(建筑体积) / (网格面积 × 参考高度)

```python
def calculate_SCD(buildings_df, reference_height=100):  # 参考高度设为100米
    total_building_volume = (buildings_df['building_area'] * buildings_df['Elevation']).sum()
    grid_volume = buildings_df['F_AREA'].iloc[0] * reference_height
    return total_building_volume / grid_volume

# 按用地类型计算SCD
scd_by_type = buildings_df.groupby('Level1_cn').apply(calculate_SCD).round(4)
print("\n各用地类型空间紧凑度:")
print(scd_by_type)
```
