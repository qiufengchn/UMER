````python
def extract_building_features(gdf):
    """提取建筑物特征"""
    # 基本特征
    features = pd.DataFrame()
    features['height'] = gdf['Height']
    features['area'] = gdf['area_m2']
    
    # 几何特征
    features['compactness'] = gdf.geometry.apply(lambda x: x.area / x.boundary.length)
    features['elongation'] = gdf.geometry.apply(lambda x: x.minimum_rotated_rectangle.length / x.minimum_rotated_rectangle.width)
    
    # 标准化
    scaler = StandardScaler()
    features_scaled = scaler.fit_transform(features)
    
    return pd.DataFrame(features_scaled, columns=features.columns)

# 提取特征
building_features = extract_building_features(data)
print("特征提取完成，形状：", building_features.shape)
````

# 基于建筑物质心的空间分析

## 1. 首先扩展质心提取代码，添加更多空间特征：



````python
import pandas as pd
import numpy as np
import geopandas as gpd
from scipy.spatial import KDTree
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

def extract_enhanced_centroids(gdf):
    """提取增强的质心特征"""
    # 基础质心信息
    centroids = pd.DataFrame({
        'building_id': range(len(gdf)),
        'x': gdf.geometry.centroid.x,
        'y': gdf.geometry.centroid.y,
        'height': gdf['Height'],
        'area': gdf['area_m2'],
        'geometry': gdf.geometry.centroid
    })
    
    # 转换为GeoDataFrame以便空间分析
    centroids_gdf = gpd.GeoDataFrame(centroids, geometry='geometry')
    
    return centroids_gdf
````



## 2. 计算建筑密度和聚集度：



````python
def calculate_spatial_metrics(gdf, radius=100):
    """计算空间指标"""
    # 创建KD树用于快速邻近搜索
    coords = np.vstack((gdf.geometry.x, gdf.geometry.y)).T
    tree = KDTree(coords)
    
    # 计算每个点在指定半径内的邻居数
    neighbors_count = tree.query_ball_point(coords, radius)
    density = [len(neighbors) for neighbors in neighbors_count]
    
    # 计算建筑密度（单位面积内的建筑数量）
    area = np.pi * radius**2
    building_density = [count/area for count in density]
    
    # 添加到GeoDataFrame
    gdf['neighbor_count'] = density
    gdf['building_density'] = building_density
    
    return gdf
````



## 3. 进行空间聚类分析：



````python
def spatial_clustering(gdf, eps=50, min_samples=5):
    """使用DBSCAN进行空间聚类"""
    # 准备聚类数据
    X = np.vstack((gdf.geometry.x, gdf.geometry.y)).T
    
    # 执行DBSCAN聚类
    clustering = DBSCAN(eps=eps, min_samples=min_samples).fit(X)
    
    # 添加聚类标签
    gdf['cluster'] = clustering.labels_
    
    return gdf
````



## 4. 执行分析并可视化：



````python
# 执行完整分析流程
centroids_gdf = extract_enhanced_centroids(data)
centroids_with_metrics = calculate_spatial_metrics(centroids_gdf, radius=100)
clustered_centroids = spatial_clustering(centroids_with_metrics, eps=50, min_samples=5)

# 创建多子图可视化
fig, axs = plt.subplots(2, 2, figsize=(20, 20))

# 1. 建筑物分布
axs[0,0].scatter(clustered_centroids.geometry.x, clustered_centroids.geometry.y, 
                 c=clustered_centroids['height'], cmap='viridis', alpha=0.6)
axs[0,0].set_title('建筑物高度分布')

# 2. 建筑密度热图
axs[0,1].scatter(clustered_centroids.geometry.x, clustered_centroids.geometry.y,
                 c=clustered_centroids['building_density'], cmap='RdYlBu', alpha=0.6)
axs[0,1].set_title('建筑密度分布')

# 3. 空间聚类结果
axs[1,0].scatter(clustered_centroids.geometry.x, clustered_centroids.geometry.y,
                 c=clustered_centroids['cluster'], cmap='tab20', alpha=0.6)
axs[1,0].set_title('空间聚类结果')

# 4. 邻居数量分布
axs[1,1].scatter(clustered_centroids.geometry.x, clustered_centroids.geometry.y,
                 c=clustered_centroids['neighbor_count'], cmap='YlOrRd', alpha=0.6)
axs[1,1].set_title('建筑物邻居数量')

# 添加颜色条
for ax in axs.flat:
    plt.colorbar(ax.collections[0], ax=ax)
    ax.set_aspect('equal')

plt.tight_layout()
plt.show()

# 输出统计信息
print("\n聚类统计：")
print(clustered_centroids.groupby('cluster').agg({
    'height': 'mean',
    'area': 'mean',
    'building_density': 'mean',
    'neighbor_count': 'mean'
}).round(2))
````
