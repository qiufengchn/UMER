```python
import geopandas as gpd
import pandas as pd
import os

# 读取 Shapefile
shp_path = r"F:\ScientificDatabase\paper03\wuhan_district\Wuhan_Subdistrict_Administrative_Boundaries_edited\Wuhan_district_detail.shp"
gdf = gpd.read_file(shp_path, encoding='utf-8')

# 定义提取区域名称的函数
def extract_district_name(path):
    # 根据路径分隔符分割字符串
    parts = path.split('/')
    # 如果没有分隔符，尝试使用其他可能的分隔符
    if len(parts) == 1:
        parts = path.split('\\')
    
    # 遍历路径部分，查找包含"区"字的部分
    for part in parts:
        if '区' in part:
            # 提取区名（去除可能的额外字符）
            district = part.split('区')[0] + '区'
            return district
    return '未知区域'  # 如果没有找到区名，返回默认值

# 添加新的区域名称字段
gdf['NAME_distr'] = gdf['path'].apply(extract_district_name)

# 查看结果
print("提取的区域名称统计：")
print(gdf['NAME_distr'].value_counts())

# 保存结果到新的 Shapefile
output_schema = {
    'NAME_distr': 'str'
}

output_path = os.path.join(os.path.dirname(shp_path), "Wuhan_district_with_names.shp")
gdf.to_file(output_path, encoding='utf-8')

# 导出区域统计到Excel
excel_path = os.path.join(os.path.dirname(shp_path), "district_names_statistics.xlsx")
district_stats = pd.DataFrame({
    '区域名称': gdf['NAME_distr'].value_counts().index,
    '街道数量': gdf['NAME_distr'].value_counts().values
})
district_stats.to_excel(excel_path, index=False, encoding='utf-8')

print("\n数据已保存到：")
print(f"1. Shapefile: {output_path}")
print(f"2. 统计表: {excel_path}")
```

# 形状复杂度指数（Compactness Index）详解

形状复杂度指数是用来衡量区域形状规则程度的一个重要指标。我们在代码中使用的计算公式是：



```python
compactness = (4 * π * 面积) / (周长²)
```



### 原理解释：

1. **理论基础**：
   - 这个指数基于圆形是所有形状中周长最短的原理
   - 当面积相同时，圆形的周长最小
   - 指数值范围在 0 到 1 之间

2. **计算特点**：
   - 值越接近 1，形状越接近圆形（最规则）
   - 值越接近 0，形状越不规则
   - 完美的圆形其指数值 = 1

### 代码示例：



```python
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np

def calculate_compactness(gdf):
    """计算并可视化形状复杂度指数"""
    
    # 计算复杂度指数
    gdf['compactness'] = (4 * np.pi * gdf.geometry.area) / (gdf.geometry.length ** 2)
    
    # 创建可视化
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    
    # 绘制复杂度分布地图
    gdf.plot(column='compactness', 
            ax=ax1,
            legend=True,
            cmap='RdYlBu',
            legend_kwds={'label': '形状复杂度指数'})
    ax1.set_title('区域形状复杂度分布')
    ax1.axis('off')
    
    # 绘制直方图
    gdf['compactness'].hist(ax=ax2, bins=20)
    ax2.set_title('形状复杂度指数分布直方图')
    ax2.set_xlabel('复杂度指数值')
    ax2.set_ylabel('频数')
    
    plt.tight_layout()
    return gdf['compactness'].describe()

# 示例使用
stats = calculate_compactness(gdf)
print("\n形状复杂度统计信息：")
print(stats)
```

### 解读标准：

- `> 0.8`: 非常规则，接近圆形
- `0.6-0.8`: 较为规则
- `0.4-0.6`: 一般规则性
- `0.2-0.4`: 较不规则
- `< 0.2`: 非常不规则

这个指标在城市规划和地理空间分析中很有价值，可以帮助识别需要优化的区域边界。
